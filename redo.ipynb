```python
df['Age'].fillna(df['Age'].mean(),inplace=True)
```

```python
#fill age column
df['Age'].isna().sum()
```

```python
#Drop cabin column
df.drop('Cabin',axis=1,inplace=True)
```

```python
df.head()
```

```python
#Check for the non-numeric column
df.info()
```

```python
df.dtypes
```

```python
#convert sex column to numerical values

gender=pd.get_dummies(df['Sex'],drop_first=True)
```

```python
df['Gender']=gender
```

```python
df.head()
```

```python
#drop the columns which are not required
df.drop(['Name','Sex','Ticket','Embarked'],axis=1,inplace=True)
```

```python
df.head()
```

```python
#Seperate Dependent and Independent variables
x=df[['PassengerId','Pclass','Age','SibSp','Parch','Fare','Gender']]
y=df['Survived']
```

```python
x
```

```python
y
```

```python
#import train test split method
from sklearn.model_selection import train_test_split
```

```python
#train test split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
#import Logistic  Regression
from sklearn.linear_model import LogisticRegression
```

```python
#Fit  Logistic Regression 
lr=LogisticRegression()
```

```python

from sklearn.model_selection import GridSearchCV
Parameter={'penalty':['l1','l2','elasticent'],'C':[1,2,3,4,5,6,10,20,30,40,50],'max_iter':[100,200,300]}
```

```python
lr=GridSearchCV(lr,param_grid=Parameter,scoring='accuracy',cv=5)
```

```python
lr.fit(x_train,y_train)
```

```python
print(lr.best_params_)
```

```python
print(lr.best_score_)
```

```python
#predict
predict=lr.predict(x_test)
```

```python
##accuracy score
from sklearn.metrics import accuracy_score,classification_report
```

```python
score=accuracy_score(y_test,predict)
print(score)
```

```python
print(classification_report(y_test,predict))
```
